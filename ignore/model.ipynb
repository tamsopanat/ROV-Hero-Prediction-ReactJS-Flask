```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
```

```python
data = pd.read_csv("./flask-server/staticFiles/rovMatchData.csv").drop(columns=['result_Blue'])
data
```

```python
allHero = []
for i in data.values:
    allHero += list(i)
allHero = list(set(allHero))
```

```python
le = LabelEncoder()
encode = le.fit_transform(allHero)
decode = le.inverse_transform(encode)
pd.DataFrame({'heroName' : allHero,
             'heroName_encode' : encode,
             'heroName_decode' : decode}).sort_values(by=['heroName_encode'])
```

```python
data_encode = data.apply(le.transform)
```

```python
X = data_encode[['carryTeamBlue', 'carryTeamRed', 'jungleTeamRed', 'supTeamRed']]
y = data_encode.supTeamBlue
```

```python
dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, y)
```

```python
input = ['Violet', 'Lindis', 'Aoi', 'Baldum']
predict = le.inverse_transform(dtree.predict(le.transform(input).reshape(1,-1)))
print(predict[0])
# predic_probat = le.inverse_transform(dtree.predict(le.transform(input).reshape(1,-1)))
```
