```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import os
```

```python
#Set directory
path = './chromedriver_mac_arm64/' #กำหนด folder ที่จะให้ระบบทำงาน
os.chdir(path) #รัน Code เพื่อสั่งให้ระบบทำงานบน folder ที่ต้องการ
```

```python
#Open webdriver 
driver = webdriver.Chrome(path+"chromedriver.exe")
driver.get('https://rov.in.th/hero')
```

```python
#Get location of hero img
heroImg = pd.DataFrame(columns=['heroName', 'imgURL'])
elements = driver.find_elements(By.CLASS_NAME, 'heroList')
for e in elements:
    heroName = e.text
    img_url = e.get_attribute('innerHTML').split('src="')[1].split('"')[0]
    print(f'''heroName : {heroName}
    img url : {img_url}''')
    temp = pd.DataFrame({'heroName':[heroName],
                        'imgURL':[img_url]})
    heroImg = pd.concat([heroImg,temp])
```

```python
heroImg.to_csv("/Users/tamsopanat/Desktop/rovHero/flask-server/staticFiles/heroImg.csv", index=False)
print("Update IMG URL Succesfully!")
```
